## 1. Introduction to Systems of Equations

### Core Concept

A system of equations is a set of two or more equations with the same variables that are considered simultaneously. The solution to a system is the set of values that satisfies all equations in the system at the same time.

### Key Ideas

* Multiple equations: Usually 2 or 3 equations with the same variables

* Same variables: Typically x and y (or x, y, z for 3 equations)

* Simultaneous solution: Looking for values that work in ALL equations

* Geometric interpretation: Each equation represents a line (or curve); the solution is where they intersect

### Types of Solutions

* One Solution: The lines intersect at exactly one point

* No Solution: The lines are parallel and never intersect

* Infinitely Many Solutions: The lines are identical (coincident)

### Basic Examples
#### Example 1: One Solution

System:
2x + y = 7
x - y = -1

Solution: x = 2, y = 3
Check: 2(2) + 3 = 7 ‚úì and 2 - 3 = -1 ‚úì

#### Example 2: No Solution

System:
y = 2x + 1
y = 2x + 5

Solution: No solution (parallel lines)

#### Example 3: Infinite Solutions

System:
y = 2x + 1
2y = 4x + 2

Solution: Infinitely many (same line)



### Python Examples
#### Example 1: Solving Systems with NumPy

```python
import numpy as np

def solve_system_2x2():
    """Solve a 2x2 system of equations"""
    print("SOLVING 2x2 SYSTEM OF EQUATIONS")
    print("=" * 40)
    
    # System: 2x + y = 7
    #         x - y = -1
    # In matrix form: AX = B
    
    # Coefficient matrix A
    A = np.array([[2, 1],   # Coefficients of 1st equation: 2x + 1y
                  [1, -1]]) # Coefficients of 2nd equation: 1x - 1y
    
    # Constant matrix B  
    B = np.array([7, -1])   # Constants: 7 and -1
    
    print(f"Coefficient matrix A:\n{A}")
    print(f"Constant matrix B: {B}")
    
    # Solve: X = A‚Åª¬πB
    try:
        solution = np.linalg.solve(A, B)
        x, y = solution
        print(f"\nSolution: x = {x}, y = {y}")
        
        # Verify the solution
        print("\nVERIFICATION:")
        eq1 = 2*x + y
        eq2 = x - y
        print(f"Equation 1: 2({x}) + {y} = {eq1} (should be 7)")
        print(f"Equation 2: {x} - {y} = {eq2} (should be -1)")
        
    except np.linalg.LinAlgError:
        print("No unique solution exists (either no solution or infinite solutions)")

solve_system_2x2()
```

#### Example 2: Graphical Solution Visualization

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_system_2x2():
    """Plot a system of 2 equations and find their intersection"""
    
    # System: 
    # Equation 1: 2x + y = 7  ‚Üí y = 7 - 2x
    # Equation 2: x - y = -1  ‚Üí y = x + 1
    
    plt.figure(figsize=(10, 8))
    
    # Generate x values
    x = np.linspace(-2, 5, 100)
    
    # Calculate y values for each equation
    y1 = 7 - 2*x  # From 2x + y = 7
    y2 = x + 1    # From x - y = -1
    
    # Plot both lines
    plt.plot(x, y1, 'b-', linewidth=2, label='2x + y = 7')
    plt.plot(x, y2, 'r-', linewidth=2, label='x - y = -1')
    
    # Find intersection point (solution)
    # Solve: 7 - 2x = x + 1 ‚Üí 6 = 3x ‚Üí x = 2
    x_sol = 2
    y_sol = x_sol + 1  # 3
    
    # Plot intersection point
    plt.plot(x_sol, y_sol, 'go', markersize=10, label=f'Solution: ({x_sol}, {y_sol})')
    
    # Add coordinate grid
    plt.axhline(y=0, color='black', linewidth=0.5)
    plt.axvline(x=0, color='black', linewidth=0.5)
    plt.grid(True, alpha=0.3)
    
    plt.title('System of Equations: Finding the Intersection', fontsize=14, fontweight='bold')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.axis('equal')
    plt.show()
    
    print("GRAPHICAL SOLUTION:")
    print(f"The lines intersect at ({x_sol}, {y_sol})")
    print(f"This means x = {x_sol}, y = {y_sol} satisfies both equations")

plot_system_2x2()
```

#### Example 3: Different Types of Systems

```python
import matplotlib.pyplot as plt
import numpy as np

def compare_system_types():
    """Compare systems with one solution, no solution, and infinite solutions"""
    
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
    
    x = np.linspace(-5, 5, 100)
    
    # System 1: One Solution
    ax1.plot(x, 2*x + 1, 'b-', linewidth=2, label='y = 2x + 1')
    ax1.plot(x, -x + 4, 'r-', linewidth=2, label='y = -x + 4')
    ax1.plot(1, 3, 'go', markersize=8, label='Solution: (1, 3)')
    ax1.set_title('One Solution\n(Lines intersect)', fontweight='bold')
    ax1.grid(True, alpha=0.3)
    ax1.legend()
    
    # System 2: No Solution
    ax2.plot(x, 2*x + 1, 'b-', linewidth=2, label='y = 2x + 1')
    ax2.plot(x, 2*x + 5, 'r-', linewidth=2, label='y = 2x + 5')
    ax2.set_title('No Solution\n(Parallel lines)', fontweight='bold')
    ax2.grid(True, alpha=0.3)
    ax2.legend()
    
    # System 3: Infinite Solutions
    ax3.plot(x, 2*x + 1, 'b-', linewidth=2, label='y = 2x + 1')
    ax3.plot(x, 2*x + 1, 'r--', linewidth=2, label='2y = 4x + 2')
    ax3.set_title('Infinite Solutions\n(Identical lines)', fontweight='bold')
    ax3.grid(True, alpha=0.3)
    ax3.legend()
    
    plt.tight_layout()
    plt.show()
    
    print("TYPES OF SOLUTIONS:")
    print("1. ONE SOLUTION: Lines intersect at one point")
    print("2. NO SOLUTION: Lines are parallel, never intersect") 
    print("3. INFINITE SOLUTIONS: Lines are identical, all points are solutions")

compare_system_types()
```

#### Example 4: Real-World Application - Cost vs Revenue

```python
import matplotlib.pyplot as plt
import numpy as np

def business_break_even():
    """Real-world example: Finding break-even point"""
    
    # Scenario: A company has fixed costs and variable costs
    # Cost equation: C = 500 + 25x  (fixed cost + cost per item)
    # Revenue equation: R = 50x     (price per item)
    # Break-even: Cost = Revenue
    
    plt.figure(figsize=(10, 8))
    
    x = np.linspace(0, 50, 100)  # Number of items
    cost = 500 + 25*x
    revenue = 50*x
    
    # Plot both lines
    plt.plot(x, cost, 'r-', linewidth=2, label='Cost: C = 500 + 25x')
    plt.plot(x, revenue, 'b-', linewidth=2, label='Revenue: R = 50x')
    
    # Find break-even point (solve: 500 + 25x = 50x)
    # 500 = 25x ‚Üí x = 20
    break_even_x = 20
    break_even_y = 50 * break_even_x  # 1000
    
    plt.plot(break_even_x, break_even_y, 'go', markersize=10, 
             label=f'Break-even: ({break_even_x}, {break_even_y})')
    
    # Shade profit and loss regions
    plt.fill_between(x, cost, revenue, where=(revenue > cost), 
                     alpha=0.2, color='green', label='Profit Region')
    plt.fill_between(x, cost, revenue, where=(revenue <= cost), 
                     alpha=0.2, color='red', label='Loss Region')
    
    plt.title('Business Break-Even Analysis', fontsize=14, fontweight='bold')
    plt.xlabel('Number of Items')
    plt.ylabel('Dollars ($)')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.show()
    
    print("REAL-WORLD SYSTEM OF EQUATIONS:")
    print("Cost equation: C = 500 + 25x")
    print("Revenue equation: R = 50x")
    print("Break-even point: Solve 500 + 25x = 50x")
    print(f"Solution: x = {break_even_x} items")
    print(f"At {break_even_x} items, both cost and revenue are ${break_even_y}")
    print("Below 20 items: Loss | Above 20 items: Profit")

business_break_even()
```

## 2. Solving Systems of Equations - Methods
### Core Concept

Different algebraic techniques to find the solution(s) to systems of equations.

### Substitution Method

* Process: Solve one equation for one variable, then substitute into the other equation(s)

* Best for: When one variable is already isolated or easily isolated

* Example:

    y = 2x + 1

    3x + 2y = 11

    Substitute: 3x + 2(2x + 1) = 11

### Elimination Method

* Process: Add or subtract equations to eliminate one variable

* Best for: When coefficients are easily aligned

* Example:

    2x + 3y = 13

    4x - 3y = 5

    Add: 6x = 18 ‚Üí x = 3

### When to Use Each Method

* Substitution: One variable is already isolated, or equations are simple

* Elimination: Coefficients are easily made opposites, or fractions are involved

* Both methods yield the same solution for consistent systems



### Python Example
#### Method Comparison Visualizer

```python
import matplotlib.pyplot as plt
import numpy as np

def compare_methods_visual():
    """Visual comparison of substitution vs elimination methods"""
    
    # System: 2x + y = 7, x - y = -1
    x_vals = np.linspace(-1, 5, 100)
    y1 = 7 - 2*x_vals  # From 2x + y = 7
    y2 = x_vals + 1    # From x - y = -1
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Plot 1: Substitution Method Visualization
    ax1.plot(x_vals, y1, 'b-', linewidth=2, label='2x + y = 7')
    ax1.plot(x_vals, y2, 'r-', linewidth=2, label='x - y = -1')
    
    # Show substitution steps
    ax1.axvline(x=2, color='green', linestyle='--', alpha=0.7, label='x = 2 (from substitution)')
    ax1.axhline(y=3, color='purple', linestyle='--', alpha=0.7, label='y = 3 (from substitution)')
    ax1.plot(2, 3, 'go', markersize=10, label='Solution: (2, 3)')
    
    ax1.set_title('Substitution Method', fontweight='bold')
    ax1.set_xlabel('x')
    ax1.set_ylabel('y')
    ax1.grid(True, alpha=0.3)
    ax1.legend()
    
    # Plot 2: Elimination Method Visualization
    ax2.plot(x_vals, y1, 'b-', linewidth=2, label='2x + y = 7')
    ax2.plot(x_vals, y2, 'r-', linewidth=2, label='x - y = -1')
    
    # Show elimination - the sum of equations
    y_sum = (y1 + y2) / 3  # Normalized for display: (2x+y)+(x-y)=3x ‚Üí x=2
    ax2.plot(x_vals, y_sum, 'g--', linewidth=2, label='Sum: 3x = 6 ‚Üí x = 2')
    
    ax2.plot(2, 3, 'go', markersize=10, label='Solution: (2, 3)')
    
    ax2.set_title('Elimination Method', fontweight='bold')
    ax2.set_xlabel('x')
    ax2.set_ylabel('y')
    ax2.grid(True, alpha=0.3)
    ax2.legend()
    
    plt.tight_layout()
    plt.show()
    
    # Print the algebraic steps
    print("SUBSTITUTION METHOD STEPS:")
    print("1. From x - y = -1 ‚Üí y = x + 1")
    print("2. Substitute: 2x + (x + 1) = 7")
    print("3. 3x + 1 = 7 ‚Üí 3x = 6 ‚Üí x = 2")
    print("4. y = 2 + 1 = 3")
    print("Solution: (2, 3)")
    
    print("\nELIMINATION METHOD STEPS:")
    print("1. Original: 2x + y = 7")
    print("2. Original: x - y = -1") 
    print("3. Add equations: (2x + y) + (x - y) = 7 + (-1)")
    print("4. 3x = 6 ‚Üí x = 2")
    print("5. Substitute: 2(2) + y = 7 ‚Üí 4 + y = 7 ‚Üí y = 3")
    print("Solution: (2, 3)")

compare_methods_visual()
```

## 3. Systems Analysis - Solutions and Equivalence
### Core Concept

Understanding the nature and types of solutions in systems of equations.

### Number of Solutions

1. One Solution: Lines intersect at one point (consistent and independent)

2. No Solution: Parallel lines never intersect (inconsistent)

3. Infinite Solutions: Same line, all points are solutions (consistent and dependent)

### Equivalent Systems

* Definition: Systems that have the exact same solution set

* Created by:

    * Multiplying an equation by a non-zero constant

    * Adding a multiple of one equation to another

    * Switching the order of equations

* Purpose: Transform complex systems into simpler equivalent systems that are easier to solve

Relationship: Equivalent systems always have the same number and type of solutions.

### Python Example
#### Solution Types and Equivalent Systems

```python
import numpy as np

def analyze_solution_types():
    """Analyze different types of solutions and equivalent systems"""
    
    print("SOLUTION TYPES AND EQUIVALENT SYSTEMS")
    print("=" * 50)
    
    # One Solution
    print("1. ONE SOLUTION:")
    A1 = np.array([[2, 1], [1, -1]])
    B1 = np.array([7, -1])
    try:
        sol1 = np.linalg.solve(A1, B1)
        print(f"System: 2x + y = 7, x - y = -1 ‚Üí Solution: {sol1}")
        
        # Equivalent system (multiply first equation by 2)
        A1_eq = np.array([[4, 2], [1, -1]])
        B1_eq = np.array([14, -1])
        sol1_eq = np.linalg.solve(A1_eq, B1_eq)
        print(f"Equivalent: 4x + 2y = 14, x - y = -1 ‚Üí Solution: {sol1_eq}")
        print("Same solution! ‚úì")
    except:
        print("No unique solution")
    
    # No Solution
    print("\n2. NO SOLUTION:")
    A2 = np.array([[2, 1], [2, 1]])
    B2 = np.array([7, 5])
    try:
        sol2 = np.linalg.solve(A2, B2)
        print(f"Solution: {sol2}")
    except np.linalg.LinAlgError:
        print("System: 2x + y = 7, 2x + y = 5 ‚Üí No solution (parallel lines)")
    
    # Infinite Solutions  
    print("\n3. INFINITE SOLUTIONS:")
    A3 = np.array([[2, 1], [4, 2]])
    B3 = np.array([7, 14])
    try:
        sol3 = np.linalg.solve(A3, B3)
        print(f"Solution: {sol3}")
    except np.linalg.LinAlgError:
        print("System: 2x + y = 7, 4x + 2y = 14 ‚Üí Infinite solutions (same line)")

analyze_solution_types()
```

## 4. Applications - Word Problems
### Core Concept

Translating real-world situations into systems of equations and solving them.

### Problem-Solving Strategy

* Define variables: Identify what each variable represents

* Write equations: Translate the word problem into mathematical equations

* Solve the system: Use substitution, elimination, or graphing

* Interpret results: Relate the mathematical solution back to the real-world context

### Common Types

* Mixture problems: Combining quantities with different properties

* Rate problems: Distance = rate √ó time scenarios

* Cost/revenue problems: Break-even analysis

* Number problems: Finding unknown numbers given relationships



### Python Example
#### Word Problem Solver

```python
def word_problem_solver():
    """Solve various types of word problems"""
    
    print("SYSTEMS OF EQUATIONS WORD PROBLEMS")
    print("=" * 50)
    
    # Problem 1: Number problem
    print("1. NUMBER PROBLEM:")
    print("The sum of two numbers is 25. Their difference is 7. Find the numbers.")
    # Equations: x + y = 25, x - y = 7
    A = np.array([[1, 1], [1, -1]])
    B = np.array([25, 7])
    solution = np.linalg.solve(A, B)
    print(f"Numbers: {solution[0]:.0f} and {solution[1]:.0f}")
    
    # Problem 2: Mixture problem
    print("\n2. MIXTURE PROBLEM:")
    print("A 20% acid solution is mixed with a 50% acid solution to get 100ml of 35% acid.")
    # Equations: x + y = 100, 0.2x + 0.5y = 35
    A = np.array([[1, 1], [0.2, 0.5]])
    B = np.array([100, 35])
    solution = np.linalg.solve(A, B)
    print(f"20% solution: {solution[0]:.0f}ml, 50% solution: {solution[1]:.0f}ml")
    
    # Problem 3: Cost problem
    print("\n3. COST PROBLEM:")
    print("3 pens and 2 pencils cost $6. 2 pens and 3 pencils cost $5.50.")
    # Equations: 3x + 2y = 6, 2x + 3y = 5.5
    A = np.array([[3, 2], [2, 3]])
    B = np.array([6, 5.5])
    solution = np.linalg.solve(A, B)
    print(f"Pen: ${solution[0]:.2f}, Pencil: ${solution[1]:.2f}")

word_problem_solver()
```
