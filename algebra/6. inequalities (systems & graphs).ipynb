## 1. Graphing Two-Variable Inequalities

### Core Concept

Two-variable inequalities are mathematical statements that show a relationship between two variables using inequality symbols (<, >, â‰¤, â‰¥, â‰ ) instead of equality. Graphing these inequalities shows all the possible solutions as regions on a coordinate plane rather than just lines or points.

### Key Components

1. Boundary Line: The related equation (replacing â‰¤ with =) that forms the edge of the solution region

2. Solution Region: The area containing all points that satisfy the inequality

3. Shading: Visual representation of the solution region

4. Solid vs Dashed Line:

    * Solid line (â‰¤ or â‰¥): Points on the line ARE included in the solution

    * Dashed line (< or >): Points on the line are NOT included in the solution

### Graphing Process

1. Graph the boundary line (treat the inequality as an equation)

2. Determine solid or dashed line based on the inequality symbol

3. Choose a test point (usually (0,0) if not on the line) to determine which side to shade

4. Shade the solution region

### Python Examples 

#### Example 1: Basic Inequality Graphing

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_basic_inequality():
    """Plot a basic linear inequality"""
    
    # Inequality: y â‰¤ 2x + 1
    x = np.linspace(-5, 5, 100)
    boundary_line = 2*x + 1
    
    plt.figure(figsize=(10, 8))
    
    # Plot boundary line (solid for â‰¤)
    plt.plot(x, boundary_line, 'b-', linewidth=3, label='y = 2x + 1')
    
    # Shade solution region (y â‰¤ 2x + 1 means shade below the line)
    plt.fill_between(x, boundary_line, -10, alpha=0.3, color='blue', 
                    label='y â‰¤ 2x + 1 solution region')
    
    plt.axhline(y=0, color='black', linewidth=0.5)
    plt.axvline(x=0, color='black', linewidth=0.5)
    plt.grid(True, alpha=0.3)
    plt.title('Graphing y â‰¤ 2x + 1', fontsize=14, fontweight='bold')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.axis('equal')
    plt.show()
    
    print("INEQUALITY ANALYSIS:")
    print("Inequality: y â‰¤ 2x + 1")
    print("Boundary line: y = 2x + 1 (solid line because of â‰¤)")
    print("Solution region: All points on or below the line")
    print("Test point (0,0): 0 â‰¤ 2(0) + 1 â†’ 0 â‰¤ 1 âœ“ (True, so shade that side)")

plot_basic_inequality()
```

#### Example 2: Comparing Inequality Types

```python
def compare_inequality_types():
    """Compare different inequality symbols"""
    
    print("COMPARING INEQUALITY TYPES")
    print("=" * 50)
    
    inequalities = [
        ('y < 2x + 1', 'dashed', 'below', 'Strict inequality - line not included'),
        ('y â‰¤ 2x + 1', 'solid', 'below', 'Inclusive inequality - line included'),
        ('y > 2x + 1', 'dashed', 'above', 'Strict inequality - line not included'),
        ('y â‰¥ 2x + 1', 'solid', 'above', 'Inclusive inequality - line included')
    ]
    
    for inequality, line_type, shading, description in inequalities:
        print(f"\n{inequality}:")
        print(f"  Line type: {line_type}")
        print(f"  Shading: {shading} the line")
        print(f"  Meaning: {description}")
    
    print("\nKEY RULES:")
    print("â€¢ < or > : Use DASHED line (points on line NOT included)")
    print("â€¢ â‰¤ or â‰¥ : Use SOLID line (points on line ARE included)")
    print("â€¢ Test point (0,0): Substitute to determine which side to shade")
    print("â€¢ If (0,0) is on the line, use another test point like (1,0)")

compare_inequality_types()
```

#### Example 3: Test Point Method Demonstrator

```python
def test_point_simple(inequality, test_point=(0, 0)):
    """Simple, reliable test point method"""
    
    print("TEST POINT METHOD - SIMPLE VERSION")
    print("=" * 50)
    print(f"Inequality: {inequality}")
    print(f"Test point: {test_point}")
    print()
    
    x, y = test_point
    
    # Manually handle common cases - more reliable than string parsing
    if inequality == 'y < 2x + 1':
        boundary_value = 2*x + 1
        symbol = '<'
        expr = '2x + 1'
    elif inequality == 'y â‰¥ -x + 3':
        boundary_value = -x + 3
        symbol = '>='
        expr = '-x + 3'
    elif inequality == 'y > x - 2':
        boundary_value = x - 2
        symbol = '>'
        expr = 'x - 2'
    elif inequality == 'y â‰¤ 3x - 1':
        boundary_value = 3*x - 1
        symbol = '<='
        expr = '3x - 1'
    else:
        print("Pre-programmed example - add more cases as needed")
        return
    
    print(f"Step 1: Boundary line: y = {expr}")
    
    print(f"Step 2: Test point ({x}, {y}):")
    print(f"        y = {y}, {expr} = {boundary_value}")
    
    # Check the inequality
    if symbol == '<':
        satisfies = y < boundary_value
        print(f"Step 3: Check {y} < {boundary_value} â†’ {satisfies}")
    elif symbol == '>=':
        satisfies = y >= boundary_value
        print(f"Step 3: Check {y} â‰¥ {boundary_value} â†’ {satisfies}")
    elif symbol == '>':
        satisfies = y > boundary_value
        print(f"Step 3: Check {y} > {boundary_value} â†’ {satisfies}")
    elif symbol == '<=':
        satisfies = y <= boundary_value
        print(f"Step 3: Check {y} â‰¤ {boundary_value} â†’ {satisfies}")
    
    if satisfies:
        print(f"Step 4: âœ“ (0,0) IS a solution â†’ Shade THAT side")
    else:
        print(f"Step 4: âœ— (0,0) is NOT a solution â†’ Shade OPPOSITE side")
    
    if symbol in ['<', '>']:
        print("Step 5: Use DASHED line")
    else:
        print("Step 5: Use SOLID line")

# Test the simple version
print("Example 1 - Simple Version:")
test_point_simple('y < 2x + 1')

print("\n" + "="*70 + "\n")

print("Example 2 - Simple Version:")
test_point_simple('y â‰¥ -x + 3')
```

## 2. Modeling with Linear Inequalities

### Core Concept

Modeling with linear inequalities is the process of translating real-world situations, constraints, and limitations into mathematical inequalities. Instead of exact equations, we use inequalities to represent scenarios where quantities have minimum/maximum limits, boundaries, or ranges rather than fixed values.

### Key Idea

While equations model exact relationships (A = B), inequalities model constraints and boundaries (A â‰¤ B, A > C, etc.). This makes them perfect for real-world situations where we have limits rather than exact amounts.

### When to Use Linear Inequalities
Use inequalities when you encounter these types of real-world constraints:

1. Budget constraints: "Spend at most $100"

2. Capacity limits: "No more than 50 people"

3. Minimum requirements: "At least 10 hours of study"

4. Physical boundaries: "Height must be over 48 inches"

5. Resource limitations: "Use less than 500 gallons of water"

### Modeling Process

#### Step 1: Identify Variables
Define what each variable represents:

* x = number of item A

* y = number of item B

* t = time in hours

etc.

#### Step 2: Translate Constraints
Convert word phrases into mathematical inequalities:

Word Phrase	|  Mathematical Inequality
"at most"	            â‰¤
"at least"	            â‰¥
"no more than"	        â‰¤
"no less than"	        â‰¥
"maximum"	            â‰¤
"minimum"	            â‰¥
"less than"	            <
"greater than"	        >

#### Step 3: Write the Inequality
Combine the variables and constraints into a mathematical statement.

#### Step 4: Interpret the Solution
Understand what the solution region represents in the real-world context.



### Python Example


```python
def production_model():
    """Model production constraints in a business"""
    
    print("BUSINESS PRODUCTION MODEL")
    print("=" * 50)
    print("Scenario: Factory produces two products")
    print("â€¢ Product A: 2 hours labor, 3 units material, $50 profit")
    print("â€¢ Product B: 3 hours labor, 1 unit material, $40 profit")
    print("â€¢ Available: 120 labor hours, 90 material units daily")
    print("â€¢ Must produce at least 10 of Product A")
    print()
    
    # Define variables
    print("Step 1: Define Variables")
    print("Let a = units of Product A")
    print("Let b = units of Product B")
    print()
    
    # Write inequalities
    print("Step 2: Write Inequalities")
    print("Labor constraint: 2a + 3b â‰¤ 120")
    print("Material constraint: 3a + b â‰¤ 90") 
    print("Minimum production: a â‰¥ 10")
    print("Non-negativity: a â‰¥ 0, b â‰¥ 0")
    print()
    
    # Profit function
    print("Step 3: Objective (What to maximize)")
    print("Profit = 50a + 40b")
    print()
    
    print("Step 4: Interpretation")
    print("The feasible region contains all production plans that:")
    print("â€¢ Don't exceed available labor hours")
    print("â€¢ Don't exceed available materials") 
    print("â€¢ Meet minimum production requirement")
    print("â€¢ Use logical (non-negative) quantities")
    
    # Analyze some production plans
    print("\nSAMPLE PRODUCTION PLANS:")
    plans = [
        (10, 20),  # Minimum A, moderate B
        (20, 15),  # Balanced production
        (30, 0),   # Maximum A, no B
        (10, 30),  # Heavy on B
        (25, 20)   # High production
    ]
    
    for a, b in plans:
        labor = 2*a + 3*b
        material = 3*a + b
        profit = 50*a + 40*b
        feasible = (labor <= 120) and (material <= 90) and (a >= 10)
        
        status = "FEASIBLE" if feasible else "NOT FEASIBLE"
        print(f"  Plan: {a}A + {b}B â†’ Labor: {labor}/120, Material: {material}/90 â†’ {status}")

production_model()
```
