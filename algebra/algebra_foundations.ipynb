Variables let us generalize patterns, e.g., 3x+5.

## 2. Substitution & Evaluating Expressions

Given an expression, we can ‚Äúplug in‚Äù a value for the variable.

### Example

Evaluate:

            3x+2 when x=4

Solution:

            3(4)+2 =12+2 =14

### Verification with Python

```python
import sympy as sp

x = sp.Symbol("x")
expr = 3*x + 2
expr.subs(x, 4)
```

## **3. Combining Like Terms**

‚ÄúLike terms‚Äù means terms with the same variable raised to the same power.

Examples of like terms:

* ( 3x + 5x )
* ( 7y - 2y )

Examples of unlike terms:

* ( 3x + 2y )
* ( x + x^2 )

### Example

Simplify:


            4x - 2x + 7


Solution:


            (4x - 2x) + 7 = 2x + 7


### Verification with Python

```python
# Define x as a SymPy symbol
x = sp.Symbol('x')
expr = 4*x - 2*x + 7
sp.simplify(expr)
```

## **4. Equivalent Expressions**

Two expressions are equivalent if they produce the same value for any input.

### Example

            3(x + 1) - x is equivalent to 2x + 3


Solution:


            3(x + 1) - x = 3x + 3 - x = 2x + 3


### Verification with Python

```python
expr1 = 3*(x+1) - x
expr2 = 2*x + 3
sp.simplify(expr1 - expr2)   # should return 0 because they are equivalent and their difference is zero
```
