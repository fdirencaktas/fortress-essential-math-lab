## 1. Two-variable linear equations 

### Core Concept
A two-variable linear equation is an equation that can be written in the standard form:

Ax + By = C

where:

* x and y are the variables (they represent unknown quantities).

* A, B, and C are constants (they are fixed numbers).

* A and B are not both zero. (If they were, you wouldn't have two variables!).

The word "linear" comes from the fact that the graph of such an equation on a coordinate plane is always a straight line.

A solution to a two-variable linear equation is any ordered pair (x, y) that makes the equation a true statement when you substitute the values in.

Crucial Insight: A single two-variable linear equation does not have just one solution. It has an infinite number of solutions‚Äîevery single point that lies on its line.

### Example:
Let's take the equation x + y = 5.

* Is (2, 3) a solution? Yes, because 2 + 3 = 5. ‚úÖ

* Is (1, 4) a solution? Yes, because 1 + 4 = 5. ‚úÖ

* Is (0, 5) a solution? Yes. ‚úÖ

* Is (5, 0) a solution? Yes. ‚úÖ

All these points, and infinitely many more, are solutions. If you plot them all on a graph, they form a straight line.

## Example with Python

```python
import sympy as sp

x, y = sp.symbols('x y')
equation = sp.Eq(x + y, 5)

print("Equation:", equation)
print("\nSome solutions:")

# Generate solutions by fixing x values
for x_val in range(-2, 3):
    y_solution = sp.solve(equation.subs(x, x_val), y)[0]
    print(f"x = {x_val}, y = {y_solution} ‚Üí {x_val} + {y_solution} = {x_val + y_solution}")
```

## 2. Slope
### Core Concept
Slope is a measure of the steepness and direction of a line. It describes how much a line rises or falls as you move from left to right along it.

The formula for slope is:

          change in y    rise    y‚ÇÇ - y‚ÇÅ
     m = ------------ = ----- = --------
          change in x    run     x‚ÇÇ - x‚ÇÅ

Where:

* m represents the slope

* (x‚ÇÅ, y‚ÇÅ) and (x‚ÇÇ, y‚ÇÇ) are any two distinct points on the line

* "rise" is the vertical change (y‚ÇÇ - y‚ÇÅ)

* "run" is the horizontal change (x‚ÇÇ - x‚ÇÅ)


### Types of Slope
Positive Slope ‚ÜóÔ∏è

* Line rises from left to right

* Example: m = 2

Negative Slope ‚ÜòÔ∏è

* Line falls from left to right

* Example: m = -1

Zero Slope ‚Üí

* Horizontal line

* No vertical change

* Example: m = 0

Undefined Slope ‚ÜïÔ∏è

* Vertical line

* No horizontal change (division by zero)

* Example: x = 3

### Real-World Interpretation
* Steepness: Larger absolute values of slope indicate steeper lines

* Direction: Positive slope = upward trend, Negative slope = downward trend

* Rate of Change: Slope represents how much y changes for each unit change in x

### Graph Examples with Python Matplotlib

#### Different types of slopes

```python
import matplotlib.pyplot as plt
import numpy as np

# Set up the plotting area
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Types of Slope', fontsize=16, fontweight='bold')

# Generate x values
x = np.linspace(-5, 5, 100)

# 1. Positive Slope
y_positive = 2*x + 1  # Slope = 2
ax1.plot(x, y_positive, 'b-', linewidth=2, label='y = 2x + 1')
ax1.set_title('Positive Slope (m = 2) ‚ÜóÔ∏è')
ax1.grid(True, alpha=0.3)
ax1.legend()
ax1.set_xlabel('x')
ax1.set_ylabel('y')

# 2. Negative Slope
y_negative = -1*x + 2  # Slope = -1
ax2.plot(x, y_negative, 'r-', linewidth=2, label='y = -x + 2')
ax2.set_title('Negative Slope (m = -1) ‚ÜòÔ∏è')
ax2.grid(True, alpha=0.3)
ax2.legend()
ax2.set_xlabel('x')
ax2.set_ylabel('y')

# 3. Zero Slope
y_zero = 0*x + 3  # Slope = 0
ax3.plot(x, y_zero, 'g-', linewidth=2, label='y = 3')
ax3.set_title('Zero Slope (m = 0) ‚Üí')
ax3.grid(True, alpha=0.3)
ax3.legend()
ax3.set_xlabel('x')
ax3.set_ylabel('y')

# 4. Undefined Slope (Vertical Line)
x_vertical = np.full(100, 2)  # x = 2 for all y values
y_vertical = np.linspace(-5, 5, 100)
ax4.plot(x_vertical, y_vertical, 'm-', linewidth=2, label='x = 2')
ax4.set_title('Undefined Slope ‚ÜïÔ∏è')
ax4.grid(True, alpha=0.3)
ax4.legend()
ax4.set_xlabel('x')
ax4.set_ylabel('y')

plt.tight_layout()
plt.show()
```

### Slope Calculation 

```python
import matplotlib.pyplot as plt

# Define two points
point1 = (1, 2)
point2 = (4, 5)

# Calculate slope
rise = point2[1] - point1[1]
run = point2[0] - point1[0]
slope = rise / run

print(f"Point 1: {point1}")
print(f"Point 2: {point2}")
print(f"Rise (change in y): {rise}")
print(f"Run (change in x): {run}")
print(f"Slope (rise/run): {slope}")

# Plot the points and line
plt.figure(figsize=(8, 6))
plt.plot([point1[0], point2[0]], [point1[1], point2[1]], 'bo-', linewidth=2, markersize=8)
plt.text(point1[0], point1[1], f'  ({point1[0]}, {point1[1]})', fontsize=12)
plt.text(point2[0], point2[1], f'  ({point2[0]}, {point2[1]})', fontsize=12)
plt.title(f'Slope Calculation: m = {rise}/{run} = {slope}', fontsize=14)
plt.grid(True, alpha=0.3)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## 3. x-intercepts and y-intercepts

### Core Concept
Intercepts are the points where a graph crosses the coordinate axes. They are fundamental for understanding and graphing linear equations and other functions.

### y-intercept

* Definition: The point where the graph crosses the y-axis

* Coordinates: Always has the form (0, b)

* How to find: Set x = 0 in the equation and solve for y

* Meaning: The value of the function when x is zero (the starting value)

### x-intercept

* Definition: The point where the graph crosses the x-axis

* Coordinates: Always has the form (a, 0)

* How to find: Set y = 0 in the equation and solve for x

* Meaning: The value of x when the function equals zero (the root or zero of the function)

### Key Properties

A line can have:

* One y-intercept (always)

* One x-intercept (most lines)

* No x-intercept (horizontal lines not on the x-axis)

* Infinite x-intercepts (lines lying exactly on the x-axis)

For linear equations in slope-intercept form y = mx + b:

* The y-intercept is clearly visible as b

* The x-intercept can be found by solving 0 = mx + b

### Graph Examples with Python Matplotlib

#### Intercepts Visualization

```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure
plt.figure(figsize=(10, 8))

# Define the line: y = 2x + 3
def line_func(x):
    return 2*x + 3

# Generate x values
x = np.linspace(-5, 5, 100)
y = line_func(x)

# Calculate intercepts
y_intercept = (0, line_func(0))  # Set x = 0
x_intercept = (-1.5, 0)  # Solve 0 = 2x + 3 ‚Üí x = -1.5

# Plot the line
plt.plot(x, y, 'b-', linewidth=2, label='y = 2x + 3')

# Plot intercept points
plt.plot(y_intercept[0], y_intercept[1], 'ro', markersize=8, label=f'y-intercept: (0, {y_intercept[1]})')
plt.plot(x_intercept[0], x_intercept[1], 'go', markersize=8, label=f'x-intercept: ({x_intercept[0]}, 0)')

# Add dashed lines to axes
plt.axhline(y=0, color='black', linewidth=0.5)
plt.axvline(x=0, color='black', linewidth=0.5)

# Add annotations
plt.annotate(f'y-intercept: (0, {y_intercept[1]})', 
             xy=y_intercept, xytext=(1, 6),
             arrowprops=dict(arrowstyle='->', color='red'))
plt.annotate(f'x-intercept: ({x_intercept[0]}, 0)', 
             xy=x_intercept, xytext=(-4, 2),
             arrowprops=dict(arrowstyle='->', color='green'))

plt.title('x-intercept and y-intercept', fontsize=14, fontweight='bold')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, alpha=0.3)
plt.legend()
plt.axis('equal')
plt.show()
```

#### Multiple Lines with Different Intercepts

```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure with subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Different Types of Intercepts', fontsize=16, fontweight='bold')

x = np.linspace(-6, 6, 100)

# Plot 1: Standard line with both intercepts
y1 = 0.5*x + 2
ax1.plot(x, y1, 'b-', linewidth=2)
ax1.plot(0, 2, 'ro', markersize=6, label='y-intercept: (0, 2)')
ax1.plot(-4, 0, 'go', markersize=6, label='x-intercept: (-4, 0)')
ax1.set_title('Both Intercepts')
ax1.grid(True, alpha=0.3)
ax1.axhline(y=0, color='black', linewidth=0.5)
ax1.axvline(x=0, color='black', linewidth=0.5)
ax1.legend()

# Plot 2: Horizontal line (no x-intercept)
y2 = np.full_like(x, 3)  # y = 3
ax2.plot(x, y2, 'b-', linewidth=2)
ax2.plot(0, 3, 'ro', markersize=6, label='y-intercept: (0, 3)')
ax2.set_title('Horizontal Line\n(No x-intercept)')
ax2.grid(True, alpha=0.3)
ax2.axhline(y=0, color='black', linewidth=0.5)
ax2.axvline(x=0, color='black', linewidth=0.5)
ax2.legend()

# Plot 3: Vertical line (no y-intercept)
x3 = np.full_like(x, 2)  # x = 2
ax3.plot(x3, x, 'b-', linewidth=2)
ax3.plot(2, 0, 'go', markersize=6, label='x-intercept: (2, 0)')
ax3.set_title('Vertical Line\n(No y-intercept)')
ax3.grid(True, alpha=0.3)
ax3.axhline(y=0, color='black', linewidth=0.5)
ax3.axvline(x=0, color='black', linewidth=0.5)
ax3.legend()

# Plot 4: Line through origin
y4 = 2*x  # y = 2x
ax4.plot(x, y4, 'b-', linewidth=2)
ax4.plot(0, 0, 'purple', marker='o', markersize=6, label='Both intercepts: (0, 0)')
ax4.set_title('Through Origin\n(Same intercept)')
ax4.grid(True, alpha=0.3)
ax4.axhline(y=0, color='black', linewidth=0.5)
ax4.axvline(x=0, color='black', linewidth=0.5)
ax4.legend()

plt.tight_layout()
plt.show()
```

#### Interactive Intercept Finder

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_line_with_intercepts(a, b, c):
    """
    Plot line in standard form: ax + by = c
    and show its intercepts
    """
    plt.figure(figsize=(10, 8))
    
    # Calculate intercepts
    x_intercept = (c/a, 0) if a != 0 else "Undefined"
    y_intercept = (0, c/b) if b != 0 else "Undefined"
    
    # Generate points for the line
    if b != 0:  # Not a vertical line
        x = np.linspace(-10, 10, 100)
        y = (c - a*x) / b
        plt.plot(x, y, 'b-', linewidth=2, label=f'{a}x + {b}y = {c}')
    else:  # Vertical line: x = c/a
        x_val = c/a
        y = np.linspace(-10, 10, 100)
        x = np.full_like(y, x_val)
        plt.plot(x, y, 'b-', linewidth=2, label=f'x = {x_val}')
    
    # Plot intercepts if they exist
    if x_intercept != "Undefined":
        plt.plot(x_intercept[0], x_intercept[1], 'go', markersize=8, 
                label=f'x-intercept: ({x_intercept[0]:.1f}, 0)')
    
    if y_intercept != "Undefined":
        plt.plot(y_intercept[0], y_intercept[1], 'ro', markersize=8, 
                label=f'y-intercept: (0, {y_intercept[1]:.1f})')
    
    # Setup plot
    plt.axhline(y=0, color='black', linewidth=0.5)
    plt.axvline(x=0, color='black', linewidth=0.5)
    plt.title(f'Line: {a}x + {b}y = {c}', fontsize=14, fontweight='bold')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.axis([-10, 10, -10, 10])
    plt.show()
    
    print(f"Equation: {a}x + {b}y = {c}")
    print(f"x-intercept: {x_intercept}")
    print(f"y-intercept: {y_intercept}")

# Example usage:
plot_line_with_intercepts(2, 3, 6)  # 2x + 3y = 6
```
